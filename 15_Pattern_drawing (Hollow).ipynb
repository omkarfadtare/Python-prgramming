#### For square pattern

`You should identify i and j indexes of stars you need to print`
- To print only first row of stars `(i==0)` range
- To print only last row of stars `(i==(n-1))` range
- To print only first column of stars `(j==0)` range
- To print only last column of stars `(j==(n-1))` range
- To print only middle row of stars `(i==n//2)` range
- To print only middle column of stars `(j==n//2)` range; index can not be a float value thats why we ahve taken floor division
- Diagonal starting from left `(i==j)` range
- Diagonal starting from right `((i+j)==(n-1))` range

```python
# Base pattern:

n = int(input('PLease enter number of rows: '))
for i in range(n):
    for j in range(n):
        print('*',end=' ')
    print()
```

```python
n = int(input('PLease enter number of rows: '))
for i in range(n):
    for j in range(n):
        if (j==0 or j==(n-1)):
            print('*',end=' ')
        else:
            print(' ',end= ' ')
    print()
```

```python
n = int(input('PLease enter number of rows: '))
for i in range(n):
    for j in range(n):
        if (i==n//2 or j==n//2):
            print('*',end=' ')
        else:
            print(' ',end= ' ')
    print()
```

```python
n = int(input('PLease enter number of rows: '))
for i in range(n):
    for j in range(n):
        if (i==j or i+j==(n-1)):
            print('*',end=' ')
        else:
            print(' ',end= ' ')
    print()
```

```python
n = int(input('PLease enter number of rows: '))
for i in range(n):
    for j in range(n):
        if (i==0 or j==0 or i==(n-1) or j==(n-1)):
            print('*',end=' ')
        else:
            print(' ',end= ' ')
    print()
```

#### For triangle pattern

```python
n = int(input('Please enter number of rows: '))
for i in range(n):
    for j in range(i+1):
        if i==(n-1) or j == 0 or i==j:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

```python
n = int(input('Please enter number of rows: '))
for i in range(n):
    for j in range(i,n):
        print('*',end=' ')
    print()
```

#### Hill pattern

```python
n = int(input('Please enter number of rows: '))
for i in range(n):
    for j in range(i+1,n):
        print(' ',end=' ')
    for j in range(i):
        if i==(n-1) or j==0:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    for j in range(i+1):
        if i==(n-1) or j==i:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

```python
# Diamond

n = int(input('Please enter number of rows: '))
for i in range(n-1):
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i):
        if (j==0):
            print('*',end=' ')
        else:
            print(' ',end=' ')
    for j in range(i+1):
        if (j==i):
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
for i in range(n):
    for j in range(i):
        print(' ',end=' ')
    for j in range(i,n):
        if (j==i):
            print('*',end=' ')
        else:
            print(' ',end=' ')
    for j in range(i,n-1):
        if j==(n-2):
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

```python
# Butterfly

n = int(input('Please enter number of rows: '))
for i in range(n):
    for j in range(i+1):
        print('*',end=' ')
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i+1,n-1):
        print(' ',end=' ')
    for j in range(i+1):
        if i==(n-1) and j==(n-1):
            print('',end=' ')
        else:
            print('*',end=' ')
    print()
for i in range(n-1):
    for j in range(i+1,n):
        print('*',end=' ')
    for j in range(i+1):
        print(' ',end=' ')
    for j in range(i):
        print(' ',end=' ')
    for j in range(i+1,n):
        print('*',end=' ')
    print()
```

- Just look at the value of i,j and then apply if and else condition to print space and star
- value of i and j will not be same in every pattern, you have to take care of that

```python
n = int(input('Please enter number of rows: '))
for i in range(n):
    p=1
    for j in range(i+1,n):
        print(' ',end=' ')
    for j in range(i):
        if i==(n-1) or j==0:
            print(p,end=' ')
            p+=1
        else:
            print(' ',end=' ')
    for j in range(i+1):
        if i==(n-1) or j==i:
            print(p,end=' ')
            p+=1
        else:
            print(' ',end=' ')
    print()
```
